```python
#Carregando Dados e bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_wine
```

```python
#Visualizando os dados
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df
```

```python
df_normalizar = (df - df.mean()) / df.std()
```

```python
matriz_cov = np.cov(df_normalizar.T)                                                  
```

```python
autovalores, autovetores = np.linalg.eig(matriz_cov)
```

```python
sorted_index = np.argsort(autovalores)[::-1]
sorted_autovalores = autovalores[sorted_index]
sorted_autovetores = autovetores[:, sorted_index]
```

```python
n_components = 2
autovetores_subset = sorted_autovetores[:, 0:n_components]
```

```python
df_pca = np.dot(df_normalizar, autovetores_subset)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=data.target, cmap='viridis', edgecolor='k', s=100)
plt.title('PCA - Wine Dataset sem Scikit-Learn')
plt.colorbar(label='')
plt.grid(True)
plt.show()
```
